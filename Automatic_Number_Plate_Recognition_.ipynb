```python
pytesseract.pytesseract.tesseract_cmd = r'/usr/bin/tesseract'
```

```python
plate_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_russian_plate_number.xml')
```

```python
folders = ["br.zip","eu.zip","us.zip"]
```

```python
os.makedirs("plates", exist_ok=True)
```

```python
for root, dirs, files in os.walk("plates"):
  for filename in files:
    if filename.lower().endswith(('.png','.jpg','jpeg')):
        img_path = os.path.join(root,filename)
        img = cv2.imread(img_path)
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        plates = plate_cascade.detectMultiScale(gray,1.1,4)
        for (x,y,w,h) in plates:
          plate_img = img[y:y+h,x:x+w]
          plate_gray = cv2.cvtColor(plate_img,cv2.COLOR_BGR2GRAY)
          text = pytesseract.image_to_string(plate_gray,config='--psm 8')
          print(f"[{os.path.relpath(img_path,'plates')}] Plate: {text.strip()}")
```

```python
import zipfile

for folder in folders:
    with zipfile.ZipFile(folder, 'r') as zip_ref:
        zip_ref.extractall("plates")
```
